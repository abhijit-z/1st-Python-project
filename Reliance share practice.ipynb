```python
# Now, we have our columns in the right order but there is still a problem. 
# Our date doesnâ€™t have the right format for matplotlib.
# Therefore, we need to convert our datetime to a number.

# If date column was index we would have to reset it first
# and since it is the index, we cannot manipulate it. 
# Therefore, we need to reset the index and then convert our datetime to a number.
```

```python
# Using the DateFormatter module from matplotlib, you can specify the format that you want to use for the date
# https://www.earthdatascience.org/courses/use-data-open-source-python/use-time-series-data-in-python/date-time-types-in-pandas-python/customize-dates-matplotlib-plots-python/

import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
```

```python
# reli_df.reset_index(inplace=True)
reli_df['Date'] = reli_df['Date'].map(mdates.date2num)
```

```python
reli_df.head()
```

```python
# This to indicate: 
# when close price > open price :--> green
# when close price < open price :--> red

color_list = []
i = 0
for el in reli_df['Close']:
    if el > reli_df['Open'][i]:
        color_list.append('g')
        i = i+1
    else:
        color_list.append('r')
        i=i+1
```

```python


# Creating Area graph

fig, axs = plt.subplots(2, sharex = True, gridspec_kw = {'hspace': 0}, figsize = (10,6) );
# Here axs has become an array in other words a numpy array
# gridspec_kw = {'hspace': 0} is to remove the space between 2 subplots

plt.subplots_adjust(left= 0.010, right = 0.85, wspace = 0)

# Shifting the location of yticks to right:
for ax in axs:
    ax.yaxis.tick_right()
    
axs[0].fill_between(reli_df['Date'], reli_df['Close'], color="skyblue", alpha=0.4)
axs[0].plot(reli_df['Date'], reli_df['Close'])
axs[0].yaxis.set_label_position("right")
axs[0].set_ylabel('Share Value/Price')

# Creating Bar chart

axs[1].bar(reli_df['Date'], reli_df['Volume']/1e6, width = 0.999, edgecolor = 'black', lw = 0.2, color = color_list)
# to show bigger values in millions we divided volume by 10^6 i.e. 1e6
# if width > 1 then bars overlapp

axs[1].yaxis.set_label_position("right")
axs[1].set_ylabel('Volume in millions')

axs[0].grid(lw = 0.4)  # lw is for darkness or lightness of the grid line
axs[1].grid(lw = 0.4)

#reli_df['Date'] = reli_df['Date'].map(mdates.date2num)

# axs[0].xaxis_date()

date_form = DateFormatter("%Y-%b-%d")
axs[0].xaxis.set_major_formatter(date_form)

## Ensure a major tick for each week using (interval=1) 
# ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1)
```
